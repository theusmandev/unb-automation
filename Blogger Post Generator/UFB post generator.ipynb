```python
import re
from IPython.display import HTML, display

def generate_novel_post_for_blogger_umera_ahmed(drive_url, novel_title):
    # ‚úÖ Extract Google Drive file ID
    if "id=" in drive_url:
        file_id = drive_url.split("id=")[-1]
    elif "file/d/" in drive_url:
        file_id = drive_url.split("file/d/")[1].split("/")[0]
    else:
        raise ValueError("Invalid Google Drive link")

    # ‚úÖ Create preview and download URLs
    preview_url = f"https://drive.google.com/file/d/{file_id}/preview"
    download_url = f"https://drive.google.com/uc?export=download&id={file_id}"

    # ‚úÖ Extract author name
    match = re.match(r"(.+?)\s+by\s+(.+)", novel_title, re.IGNORECASE)
    if match:
        novel_name = match.group(1).strip()
        author_name = match.group(2).strip()
    else:
        novel_name = novel_title
        author_name = "Umera Ahmed"

    # ‚úÖ SEO labels
    labels = [
        "Umera Ahmed Novels",
        "Download All Umera Ahmed Books",
        "Aab e Hayat by Umera Ahmed",
        "Umera Ahmed",
        "Best Urdu Novels",
        "Urdu Fiction Bank"
    ]

    # ‚úÖ SEO meta description
    description = f"{novel_title} is one of the most inspiring Urdu novels by Umera Ahmed, exploring love, faith, and spiritual growth."
    description = description[:150]

    # ‚úÖ Blogger HTML post content
    html_content = f"""
    <h1>{novel_title} | Download PDF</h1>

<div class="novel-post">
    <p><b>{novel_title}</b> is one of the finest works by <b>{author_name}</b> ‚Äî a legendary Urdu novelist whose writings touch the heart and soul.
    Her storytelling blends <b>faith, love, spirituality, and human emotions</b> in a way that inspires readers and transforms perspectives.</p>

    <p>{author_name}'s novels are deeply rooted in real-life experiences, exploring how people struggle between <b>faith and materialism</b>,
    and how belief in Allah can change one‚Äôs destiny. Her stories carry emotional depth and moral lessons that make them timeless.</p>

    <p>Some of her most acclaimed works include <i>Pir-e-Kamil</i>, <i>La Hasil</i>, <i>Shehr-e-Zaat</i>, <i>Zindagi Gulzar Hai</i>, and <i>Aab-e-Hayat</i>.
    Each of these novels inspires readers to reconnect with spirituality, purpose, and inner peace.</p>

    <p><b>{novel_name}</b> continues the legacy of meaningful Urdu fiction ‚Äî filled with life-changing lessons, soulful characters, and a powerful connection to divine love.
    It‚Äôs a must-read for every Urdu literature lover who enjoys stories that go beyond emotions and touch the soul.</p>

    <p>Now available on <b>Urdu Novel Bank</b>, you can <b>read {novel_title} online</b> or <b>download the full PDF</b> version absolutely free.
    Step into Umera Ahmed‚Äôs world of <b>faith, love, and transformation</b> ‚Äî where every word leaves a lasting impact.</p>

    <div class="novel-details">
        <h2>üìò Book Details</h2>
        <ul>
            <li><b>Book Name:</b> {novel_name}</li>
            <li><b>Author:</b> {author_name}</li>
            <li><b>Status:</b> Completed</li>
            <li><b>Language:</b> Urdu</li>
            <li><b>Format:</b> PDF + Read Online</li>
        </ul>
    </div>

    <div class="novel-read">
        <h2>üìñ Read Online {novel_title}</h2>
        <iframe loading="lazy" src="{preview_url}"></iframe>
    </div>

    <div class="novel-download">
        <a href="{download_url}" target="_blank" rel="noopener noreferrer">üì• Download PDF</a>
    </div>
<br>
    <div class="novel-tags">
        <h3>üè∑Ô∏è Tags:</h3>
        <i>{novel_title} PDF Download, {author_name} Urdu Books, Read Online {novel_name}, Best Urdu Novels by Umera Ahmed</i>
    </div>
</div>
"""

    # ‚úÖ Display Blogger Output
    display(HTML(f"""
    <h3>üìÑ Blogger Embed Code</h3>
    <textarea id="final_code" style="width:100%;height:300px;">{html_content}</textarea>
    <button onclick="navigator.clipboard.writeText(document.getElementById('final_code').value)">üìã Copy Embed Code</button>

    <h3>üè∑Ô∏è Blogger Labels</h3>
    <textarea id="labels_code" style="width:100%;height:100px;">{', '.join(labels)}</textarea>
    <button onclick="navigator.clipboard.writeText(document.getElementById('labels_code').value)">üìã Copy Labels</button>

    <h3>üìù SEO Description</h3>
    <textarea id="description_code" style="width:100%;height:100px;">{description}</textarea>
    <button onclick="navigator.clipboard.writeText(document.getElementById('description_code').value)">üìã Copy Description</button>
    """))


# ‚úÖ Example usage
drive_link = "https://drive.google.com/file/d/1KuWwZVzzQk93sHLJEPZRWqCK3nLTxkbC/view?usp=drive_link"
novel_title = "Alif novel by Umera Ahmed"
generate_novel_post_for_blogger_umera_ahmed(drive_link, novel_title)
```
