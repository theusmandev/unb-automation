```python
import os
from google.colab import auth
from googleapiclient.discovery import build
from google.colab import files

# 1. Google Drive Authentication
auth.authenticate_user()
drive_service = build('drive', 'v3')

def get_pdf_links(folder_id):
    all_items = []
    page_token = None

    # 2. Folder se saari PDF files nikalna (Handling multiple pages)
    while True:
        query = f"'{folder_id}' in parents and mimeType='application/pdf' and trashed=false"
        results = drive_service.files().list(
            q=query,
            fields="nextPageToken, files(id, name)",
            pageToken=page_token
        ).execute()

        all_items.extend(results.get('files', []))
        page_token = results.get('nextPageToken', None)
        if not page_token:
            break

    if not all_items:
        print('Is folder mein koi PDF files nahi mili.')
        return

    # 3. Data ko Process karna
    processed_list = []
    for item in all_items:
        full_name = item['name']
        file_id = item['id']

        # Title clean-up
        title = os.path.splitext(full_name)[0] # Extension hatana
        title = title.title() # Capitalize Words

        # "By" se pehle "Novel" lagana
        if " By " in f" {title} ":
            title = title.replace(" By ", " Novel By ")

        link = f"https://drive.google.com/file/d/{file_id}/view?usp=sharing"

        # List mein save karna sorting ke liye
        processed_list.append({
            'title': title.strip(),
            'link': link
        })

    # 4. A to Z Sorting (Title ki bunyad par)
    processed_list.sort(key=lambda x: x['title'])

    # 5. Text file mein save karna
    output_filename = "sorted_pdf_list.txt"
    with open(output_filename, "w", encoding="utf-8") as f:
        for entry in processed_list:
            f.write(f"{entry['title']}\n{entry['link']}\n\n")

    print(f"Total {len(processed_list)} files mili hain. File download ho rahi hai...")
    files.download(output_filename)

# --- YAHAN APNI FOLDER ID DAALEIN ---
FOLDER_ID = '16G1WQKY-8wxHhExQdbqmjmma2Dldzg5P'

get_pdf_links(FOLDER_ID)
```
